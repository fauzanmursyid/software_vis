```python
import pandas as pd
from IPython.core.display import HTML
data = pd.read_csv("top_software_dataset.csv", engine='python')
```

```python
print (data.head(30))
print (data.iloc[0,0])
index_max = data.shape[0]
softwares_local = []
softwares_master = []
for i in range(index_max -1):
    softwares_local = data["software"].iloc[i].split(",")
    for software in softwares_local:
        if software != "" and software not in softwares_master:
            softwares_master.append(software)
        else:
            pass
print(len(softwares_master))
```

```python
dates_master = []
for i in range(index_max -1):
    date_local = str(data["month"].iloc[i]) + "/" + str(data["year"].iloc[i])
    if date_local not in dates_master:
        dates_master.append(date_local)
    else:
        pass
print(dates_master)
```

```python
temporal_DF = pd.DataFrame(index = dates_master, columns = softwares_master)
temporal_DF.fillna(0, inplace=True)
```

```python
temporal_DF
```

```python
for i in range(index_max -1):
    softwares_local = data["software"].iloc[i].split(",")
    date_local = str(data["month"].iloc[i]) + "/" + str(data["year"].iloc[i])
    for software in softwares_local:
        if software != "":
            temporal_DF[software].loc[date_local] += 1
        else:
            pass
temporal_DF
```

```python
temporal_DF = temporal_DF.sort_index()
```

```python
temporal_DF = temporal_DF.transpose()
temporal_DF.to_excel("temporal_DF_top_software_dataset.xlsx")
```
